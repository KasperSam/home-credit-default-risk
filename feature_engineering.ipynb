```python
import polars as pl

dir = "data/"
app_train = pl.read_csv(dir + "application_train.csv")
app_test = pl.read_csv(dir + "application_test.csv")
bureau = pl.read_csv(dir + "bureau.csv")
bureau_balance = pl.read_csv(dir + "bureau_balance.csv")
prev_app = pl.read_csv(dir + "previous_application.csv")
pos_cash = pl.read_csv(dir + "POS_CASH_balance.csv")
installments = pl.read_csv(dir + "installments_payments.csv")
credit_card = pl.read_csv(dir + "credit_card_balance.csv")
```

```python

def create_features(app, bureau, bureau_balance, prev_app, pos_cash, installments, credit_card):

    # ====================== 2. Application-level Features ======================
    app = app.with_columns([
        (pl.col("AMT_CREDIT") / pl.col("AMT_INCOME_TOTAL")).alias("CREDIT_INCOME_RATIO"),
        (pl.col("AMT_ANNUITY") / pl.col("AMT_INCOME_TOTAL")).alias("ANNUITY_INCOME_RATIO"),
        (pl.col("AMT_ANNUITY") / pl.col("AMT_CREDIT")).alias("ANNUITY_CREDIT_RATIO"),
        (pl.col("AMT_GOODS_PRICE") / pl.col("AMT_CREDIT")).alias("GOODS_CREDIT_RATIO"),
        (pl.col("AMT_INCOME_TOTAL") / pl.col("CNT_FAM_MEMBERS")).alias("INCOME_PER_PERSON"),
        (-pl.col("DAYS_BIRTH") / 365).alias("AGE_YEARS"),
        (-pl.col("DAYS_EMPLOYED") / 365).alias("EMPLOYMENT_YEARS"),
        (-pl.col("DAYS_EMPLOYED") / -pl.col("DAYS_BIRTH")).alias("EMPLOYMENT_AGE_RATIO"),
    ])


    # ====================== 3. Bureau Features ======================
    # Aggregate bureau_balance first
    bb_agg = (
        bureau_balance.group_by("SK_ID_BUREAU")
        .agg([
            pl.len().alias("BB_MONTHS_COUNT"),
            pl.col("MONTHS_BALANCE").mean().alias("BB_MONTHS_MEAN"),
            pl.col("STATUS").n_unique().alias("BB_STATUS_UNIQUE")
        ])
    )

    bureau = bureau.join(bb_agg, on="SK_ID_BUREAU", how="left")

    # Aggregate bureau per SK_ID_CURR
    bureau_agg = (
        bureau.group_by("SK_ID_CURR")
        .agg([
            pl.len().alias("BUREAU_LOAN_COUNT"),
            pl.col("AMT_CREDIT_SUM").sum().alias("TOTAL_CREDIT_SUM"),
            pl.col("AMT_CREDIT_SUM_DEBT").sum().alias("TOTAL_DEBT_SUM"),
            (pl.col("AMT_CREDIT_SUM_DEBT").sum() / pl.col("AMT_CREDIT_SUM").sum()).alias("DEBT_CREDIT_RATIO"),
            pl.col("CREDIT_DAY_OVERDUE").max().alias("MAX_DAYS_OVERDUE"),
        ])
    )

    app = app.join(bureau_agg, on="SK_ID_CURR", how="left")


    # ====================== 4. Previous Applications ======================
    prev_agg = (
        prev_app.group_by("SK_ID_CURR")
        .agg([
            pl.len().alias("PREV_APP_COUNT"),
            (pl.col("AMT_APPLICATION") / pl.col("AMT_CREDIT")).mean().alias("MEAN_APP_CREDIT_RATIO"),
            pl.col("AMT_CREDIT").mean().alias("MEAN_PREV_CREDIT"),
        ])
    )

    app = app.join(prev_agg, on="SK_ID_CURR", how="left")


    # ====================== 5. Installments Payments ======================
    installments_agg = (
        installments.group_by("SK_ID_CURR")
        .agg([
            ((pl.col("AMT_PAYMENT") / pl.col("AMT_INSTALMENT")) - 1).mean().alias("PAYMENT_PERC_MEAN"),
            (pl.col("AMT_PAYMENT") - pl.col("AMT_INSTALMENT")).mean().alias("PAYMENT_DIFF_MEAN"),
            (pl.col("DAYS_ENTRY_PAYMENT") - pl.col("DAYS_INSTALMENT")).mean().alias("DAYS_LATE_MEAN")
        ])
    )

    app = app.join(installments_agg, on="SK_ID_CURR", how="left")


    # ====================== 6. POS / Cash Balance ======================
    pos_agg = (
        pos_cash.group_by("SK_ID_CURR")
        .agg([
            pl.len().alias("POS_COUNT"),
            pl.col("SK_DPD").mean().alias("POS_DPD_MEAN"),
            pl.col("SK_DPD_DEF").mean().alias("POS_DPD_DEF_MEAN"),
        ])
    )

    app = app.join(pos_agg, on="SK_ID_CURR", how="left")


    # ====================== 7. Credit Card Balance ======================
    credit_agg = (
        credit_card.group_by("SK_ID_CURR")
        .agg([
            (pl.col("AMT_BALANCE") / pl.col("AMT_CREDIT_LIMIT_ACTUAL")).mean().alias("CREDIT_UTIL_MEAN"),
            pl.col("AMT_PAYMENT_CURRENT").mean().alias("CREDIT_PAYMENT_MEAN"),
        ])
    )

    app = app.join(credit_agg, on="SK_ID_CURR", how="left")
    return app
```

```python
# Create features
app_train_final = create_features(
    app_train, bureau, bureau_balance, prev_app, pos_cash, installments, credit_card
)

app_test_final = create_features(
    app_test, bureau, bureau_balance, prev_app, pos_cash, installments, credit_card
)
```
